### üìù Questions for Reflection  

1. **How is prompt chaining similar to solving a jigsaw puzzle?**  
üëâ Solving a jigsaw puzzle requires placing pieces one by one until the full picture is revealed. Similarly, prompt chaining breaks a complex task into smaller steps, where each prompt‚Äôs output acts like a puzzle piece that guides the next step, eventually leading to the complete solution.  

2. **What‚Äôs the main difference between iteration and chaining?**  
üëâ **Iteration** involves refining the *same* prompt multiple times with small tweaks until the output is satisfactory.  
üëâ **Chaining** involves using a *sequence of different prompts*, where each one builds on the result of the previous step to accomplish a larger, multi-stage task.  

3. **Why might prompt chaining be more effective than asking AI one big question?**  
üëâ Breaking a complex request into smaller, focused prompts helps the AI remain accurate, reduces confusion, and gives you more control over each step. This usually produces clearer, more structured, and higher-quality results compared to asking one long, overloaded question.  

4. **How can the context window of an AI model affect your chaining strategy?**  
üëâ The **context window** is the maximum amount of text the AI can ‚Äúsee‚Äù and remember at once.  
- If the context window is **large**, the AI can retain more of the earlier steps, making chaining smoother.  
- If the context window is **small**, you may need to summarize or carry forward only the most important outputs, so the AI doesn‚Äôt lose track of essential information.

## <span style="color:#DAA520;">1. Use Case</span>

I chose to build a **Crisis Helper Bot**.  

> **Why this use case?**  
During emergencies ‚Äî such as **panic attacks, natural disasters, or unsafe situations** ‚Äî people often don‚Äôt know what to do immediately.  
A chatbot can provide **calm, step-by-step guidance** until professional help arrives.

**Target Users:**  
- People facing sudden **mental health crises** (panic, anxiety)  
- Victims of **domestic violence** seeking discreet help  
- Individuals caught in **natural disasters or accidents**  
- Students or workers under extreme stress  

**Expected Benefits:**  
- Instant, judgment-free support  
- Provides **grounding techniques** and safety instructions  
- Available **24/7**, unlike human hotlines  

**Limitations:**  
- Cannot replace real emergency services  
- Limited to text-based guidance  
- May oversimplify complex situations  

---

## <span style="color:#DAA520;">2. Justification</span>

**Why is this problem worth solving?**  
Emergencies need **immediate response**, but panic often clouds judgment.  
This bot provides **calm, structured steps** to regain control.  

**Why a chatbot instead of apps/humans?**  
- Faster than browsing websites in crisis  
- More available than hotlines (which can be busy)  
- **Anonymous & private**, important for abuse victims  

**Specific tasks the bot will perform:**  
- Teach **breathing & grounding exercises**  
- Provide **safety instructions** during disasters  
- Share **emergency numbers** and **helpline links**  
- Send **supportive, empathetic messages**  

---

## <span style="color:#DAA520;">3. LLM Choice (Groq)</span>

I chose **`llama-3.1-8b-instant`** because:

- **Real-time responses** ‚Üí optimized for low-latency inference, critical in crisis  
- **Efficient & lightweight** ‚Üí smaller than larger variants like `mixtral-8x7b`  
- **Multilingual support** ‚Üí works across many languages  
- **Simpler integration** ‚Üí designed for chat-based interaction  

> If deeper reasoning is needed, upgrading to **`mixtral-8x7b`** is an option.
